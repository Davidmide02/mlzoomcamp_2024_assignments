```python
# One-hot Encoding
# to get X for each split
 
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=False)
train_dic = df_train.to_dict(orient = 'records')
val_dic = df_val.to_dict(orient = 'records')
# fit and transform for train split 
X_train = dv.fit_transform(train_dic)
# transform only for val split
X_val = dv.transform(val_dic)

len(X_train), len(X_val)
```

```python
dv.get_feature_names_out()
```

```python
# train the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
# hard prediction becuase we already have the exact label
model.predict(X_train)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
churn_decision = (y_pred >=0.5)
```

```python
df_pred = pd.DataFrame()
df_pred['probability'] = y_pred
df_pred['prediction'] = churn_decision.astype(int)
df_pred['actual'] = y_val.values
df_pred
```

```python
df_pred['correct'] = df_pred.prediction == df_pred.actual
df_pred
```

```python
# Q4: ans
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_val)
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy of model is :", round(accuracy,2))
```

```python
# Q5:
original_accuracy = accuracy
model_feature_accuracy = {}

for feature in df_train.columns:
    df_temp_train = df_train.drop(feature, axis=1)
    df_temp_val = df_val.drop(feature, axis=1)
    dv = DictVectorizer(sparse=False)
    train_dic_temp = df_temp_train.to_dict(orient = 'records')
    val_dic_temp = df_temp_val.to_dict(orient = 'records')
    X_train_temp = dv.fit_transform(train_dic_temp)
    X_val_temp = dv.transform(val_dic_temp)
    
    # train model
    model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
    model.fit(X_train_temp, y_train)
    
    # predict with X_val split
    y_pred_temp = model.predict(X_val_temp)
    
    # calcualte feature accuracy
    accuracy_temp = accuracy_score(y_val, y_pred_temp)
    
    # get the diff btw d original and isolated feature accuracy 
    diff = original_accuracy - accuracy_temp
    
    # save the feature and the accuracy in a dictionary
    model_feature_accuracy[feature] = diff
model_feature_accuracy   
```

```python
print(min(model_feature_accuracy, key=model_feature_accuracy.get))
```

```python
# Q6: 

reg = [0.01, 0.1, 1, 10, 100]
model_regularization_accuracy = {}

for c in reg:
    
    # train model
    model = LogisticRegression(solver='liblinear', C=c, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    
    # predict with X_val split
    y_pred_c = model.predict(X_val)
    
    # calcualte feature accuracy
    accuracy_c = round(accuracy_score(y_val, y_pred_c),3)
     
    # save the feature and the accuracy in a dictionary
    model_regularization_accuracy[c] = accuracy_c
model_regularization_accuracy   
```

```python
# Ans 6:
print(max(model_regularization_accuracy, key=model_regularization_accuracy.get))
```
