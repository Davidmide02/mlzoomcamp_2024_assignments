```python
churn_partner= df_full_train[df_full_train.marital =='married']['y'].mean()
churn_partner
```

```python
# difference
# > 0  no churn
# < 0   churn

global_churn - churn_partner
```

```python
# risk
# >1 churn
# <1 no churn
churn_partner/global_churn
```

```python
df_group = df_full_train.groupby('contact').y.agg(['mean','count'])
df_group['diff'] = global_churn - df_group['mean']
df_group['risk'] = df_group['mean'] / global_churn
df_group
```
